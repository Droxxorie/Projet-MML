```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sc
```

### Parsing

```python
data = pd.read_csv("data/ValeursFoncieres-2019-S2.txt", delimiter='|', dtype= str)
data["Valeur fonciere"] = data["Valeur fonciere"].str.replace(",", ".", regex= False).astype(float) #* str to float
data["Date de mutation"] = pd.to_datetime(data["Date mutation"], format="%d/%m/%Y", errors="coerce") #* str do date_time
filtered = data.dropna(subset=["Date mutation", "Valeur fonciere"]) #* remove empty cells
# filtered = filtered[(filtered["Valeur fonciere"] < 1_000_000)] #* Remove big numbers for plot
```

```python
plt.figure()
plt.scatter(filtered["Date de mutation"], filtered["Valeur fonciere"], s= 1, alpha= 0.5, marker= '.')
plt.show()
```
